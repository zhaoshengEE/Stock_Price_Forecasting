```python
%matplotlib inline
import numpy as np                # advanced math library
import matplotlib.pyplot as plt   # MATLAB like plotting routines
import pandas as pd               # For loading the data into dataframe
```

```python
df = pd.read_csv("GME_stock.csv")
```

```python
date = pd.to_datetime(df['date'])
```

```python
close_price = df[['close_price']]
```

```python
plt.figure(figsize = (11, 7))
plt.plot(date, close_price, color= 'blue')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('GameStop close price')
plt.savefig('close_price.png')
plt.show()
```

```python
plt.figure(figsize = (11, 7))
plt.plot(date[-20:], close_price[-20:], color= 'blue')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('GameStop close price in the last 20 days')
plt.savefig('close_price_last_20_days.png')
plt.show()
```
